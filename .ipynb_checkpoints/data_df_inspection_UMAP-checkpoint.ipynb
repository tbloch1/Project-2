```python
tha = pd.read_pickle(base_f_loc+'tha.pkl')
thb = pd.read_pickle(base_f_loc+'thb.pkl')
thc = pd.read_pickle(base_f_loc+'thc.pkl')
thd = pd.read_pickle(base_f_loc+'thd.pkl')
the = pd.read_pickle(base_f_loc+'the.pkl')
```

```python
display(tha.columns[0:16])
cols = ['E_'+str(i+1) for i in range(16)]+tha.columns.tolist()[16:]
tha = tha[cols]
thb = thb[cols]
thc = thc[cols]
thd = thd[cols]
the = the[cols]
```

```python
def how_much(df):
    equa_r = df[(np.abs(df['pos_z_gsm']/6378.1)<1.0) &
                 (df['r'] >= 7) & (df['r'] <= 10)]
    dawn = equa_r[(3 <= equa_r['mlt']) &
                  (equa_r['mlt'] <= 9)]
    dusk = equa_r[(15 <= equa_r['mlt']) &
                  (equa_r['mlt'] <= 21)]
    print('DF: '+str(len(df)),'EQR: '+str(len(equa_r)),'Dawn: '+str(len(dawn)),'Dusk: '+str(len(dusk)))
    return dawn,dusk
```

```python
a_dawn, a_dusk = how_much(tha)
b_dawn, b_dusk = how_much(thb)
c_dawn, c_dusk = how_much(thc)
d_dawn, d_dusk = how_much(thd)
e_dawn, e_dusk = how_much(the)
```

```python
dawn = pd.concat([a_dawn,b_dawn,c_dawn,d_dawn,e_dawn],axis=0,
                 sort=False,ignore_index=False)
dusk = pd.concat([a_dusk,b_dusk,c_dusk,d_dusk,e_dusk],axis=0,
                 sort=False,ignore_index=False)
dawn.info()
```

# <center> UMAP </center>

```python
from umap import UMAP
from hdbscan import HDBSCAN as hdb
from sklearn.preprocessing import MinMaxScaler as mms
```

```python
dd_data = pd.concat([dawn,dusk],axis=0,sort=False)
dd_en = dd_data.iloc[:,0:11]
dd_en = dd_en.dropna()

scaler = mms().fit(dd_en)

to_red = scaler.transform(dd_en)
```

```python
reducer = UMAP(n_neighbors=30,metric='euclidean')
red_mod = reducer.fit(to_red)
red = red_mod.transform(to_red)
```

```python
clusterer = hdb(min_cluster_size=30, min_samples=50)
reclu = clusterer.fit(red)

labels = reclu.labels_
dd_en['labels'] = labels
red_df = pd.DataFrame(red,columns=['x','y'])
red_df['labels'] = labels
print(sum(labels==0))
set(labels)
```
